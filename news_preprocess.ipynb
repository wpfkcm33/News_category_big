```python
#특정 카테고리별로 동일한 수의 뉴스를 선택
import pandas as pd

df1 = pd.read_csv('NewsResult_20230101-20231231_정치.csv')
df2 = pd.read_csv('NewsResult_20230101-20231231_경제.csv')
df3 = pd.read_csv('NewsResult_20230101-20231231_사회.csv')
df4 = pd.read_csv('NewsResult_20230101-20231231_문화.csv')
df5 = pd.read_csv('NewsResult_20230101-20231231_IT_과학.csv')

balanced_categories_news_df = pd.concat([df1, df2, df3, df4, df5])
balanced_categories_news_df.to_csv('balanced_categories_news.csv', index=False, encoding='utf-8-sig')
```

```python
balanced_categories_news_df = balanced_categories_news_df[['제목', '통합 분류1', '키워드', '특성추출(가중치순 상위 50개)', '본문']]
balanced_categories_news_df.info
```

```python
random_sample_news_df = random_sample_news_df[['제목', '통합 분류1', '키워드', '특성추출(가중치순 상위 50개)', '본문']]
random_sample_news_df.info
```

```python
# 데이터프레임의 복사본 생성
balanced_categories_news_df_copy = balanced_categories_news_df.copy()

# 복사본에서 변경 수행
balanced_categories_news_df_copy['카테고리'] = balanced_categories_news_df_copy['통합 분류1'].str.split('>').str[0]

# '통합 분류1' 열 제거
balanced_categories_news_df_copy.drop('통합 분류1', axis=1, inplace=True)

# 필요하다면 원본 데이터프레임을 복사본으로 대체
balanced_categories_news_df = balanced_categories_news_df_copy
```

```python
# 데이터프레임의 복사본 생성
random_sample_news_df_copy = random_sample_news_df.copy()

# 복사본에서 변경 수행
random_sample_news_df_copy['카테고리'] = random_sample_news_df_copy['통합 분류1'].str.split('>').str[0]

# '통합 분류1' 열 제거
random_sample_news_df_copy.drop('통합 분류1', axis=1, inplace=True)

# 필요하다면 원본 데이터프레임을 복사본으로 대체
random_sample_news_df = random_sample_news_df_copy
```

### balanced_categories_news_df를 학습 데이터로 사용

```python
from sklearn.model_selection import train_test_split
# 카테고리 열을 기준으로 데이터를 7:3 비율로 나눔
train_df, test_df = train_test_split(balanced_categories_news_df, test_size=0.3, stratify=balanced_categories_news_df['카테고리'], random_state=42)
train_df.to_csv('BalancedNewsCategory_Train_2023.csv', index=False, encoding='utf-8-sig')
test_df.to_csv('BalancedNewsCategory_Test_2023.csv', index=False, encoding='utf-8-sig')
```
